```python
def hypothesis(theta, x):
    z = np.dot(x,theta)
    #print('z= ',z)
    return(1/(1+np.exp(-z)))
#correct now, dont change
```

```python
def cost(h,y):
    m=len(y)
    return ((1/m)*(np.dot(-y.T,np.log(h))-np.dot((1-y).T,np.log(1-h))))
```

```python
def gradient_descent(x,y,theta,alpha,loops):
    m=len(y)
    theta_matrix=np.zeros((14,0))
    for i in np.unique(y):
        y_multi= np.where(y==i,1,0)
        for loop in range(loops):
            h=hypothesis(theta,x)
            theta = theta - (alpha/m)*(np.dot(x.T,h-y_multi))
        theta_matrix=np.append(theta_matrix,theta,axis=1)
    return theta_matrix
```

```python
def hypothesis_final(theta,x):
    z0=np.dot(x,theta[:,0])
    z1=np.dot(x,theta[:,1])
    z2=np.dot(x,theta[:,2])
    return(1/(1+np.exp(-z0)),1/(1+np.exp(-z1)),1/(1+np.exp(-z2)))
```

```python
def fit(x,y,alpha,loops):
    theta=np.zeros((14,1))
    theta_min=gradient_descent(x,y,theta,alpha,loops)
    h0,h1,h2=hypothesis_final(theta_min,x)
    h0=h0.reshape(x.shape[0],1)
    h1=h1.reshape(x.shape[0],1)
    h2=h2.reshape(x.shape[0],1)
    H_final=np.zeros((x.shape[0],0))
    H_final=np.append(H_final,h0,axis=1)
    H_final=np.append(H_final,h1,axis=1)
    H_final=np.append(H_final,h2,axis=1)
    h_max = np.amax(H_final, axis=1,keepdims = True)
    index = np.where(H_final == h_max)
    prediction = index[1]
    prediction = prediction.reshape(x.shape[0],1)
    return prediction
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded)
X_train = np.c_[np.ones(len(X_train),dtype='int64'),X_train]
X_test = np.c_[np.ones(len(X_test),dtype='int64'),X_test]
y_train.shape
np.unique(y_test)
y1 = np.zeros([y_train.shape[0], len(np.unique(y_train))])
y1 = pd.DataFrame(y1)
```

```python
alpha = 0.005
loops = 10000
train_predictions = fit(X_train,y_train,alpha,loops)
test_predictions = fit(X_test,y_test,alpha,loops)
```

```python
print("train accuracy: {} %".format(100 - np.mean(np.abs(train_predictions - y_train)) * 100))
print("test accuracy: {} %".format(100 - np.mean(np.abs(test_predictions - y_test)) * 100))
```
